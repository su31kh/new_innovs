```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def eror_gen(N):
  np.random.seed(555)
  NxtBits = 10000
  fs_tx = np.random.randint(0,2,(1,NxtBits))
  s_tx = np.repeat(fs_tx, N)
  bsc_Noise = np.random.rand(1,N * NxtBits)

  temp = []
  for i in range(0, 50+1):
    # temp.append(eror_gen(n, i/100, 11*i))
    p = i/100
    # np.random.seed(i * 11)
    # bsc_Noise = np.random.rand(1,N * NxtBits)
    bscc_Noise = np.where(bsc_Noise > 1-p, 1, 0)

    xored_arr1 = s_tx ^ bscc_Noise

    op_listt = []

    for i in range(NxtBits):
      k = np.sum(xored_arr1[0][i*N:(i+1)*N])
      if k <= N//2:
        op_listt.append(0)
      else:
        op_listt.append(1)

    error = fs_tx ^ op_listt

    unique_element, count = np.unique(error, return_counts=True)

    try:
      temp.append((float)(count[1]) / NxtBits)
    except:
      temp.append(unique_element[0])
  return temp
  
# print(eror_gen(3))        

##################################################################

def calculate_sum(n, p):
    if n == 1:
        return p
    total = 0
    for r in range((n+1)//2, n+1):
        coefficient = math.comb(n, r) # calculate nCr
        term = coefficient * (p**r) * ((1-p)**(n-r))
        total += term
        # print(total)
    return total

# print(calculate_sum(11, 0.5))
```

```python
N_valsh = [3, 7, 11]
n_anly_plth = []
for n in N_valsh:
  n_anly_plth.append(eror_gen(n))

##################################################################

N_vals = [1, 3, 7, 11]
n_anly_plt = []
for n in N_vals:
  temp = []
  for i in range(0, 50+1):
    temp.append(calculate_sum(n, i/100))
  n_anly_plt.append(temp)
# print(n_anly_plt[2])
```

```python
# fig, plt = plt.subplots()

# Plot the data on the figure
plt.figure(figsize = (15,15))

plt.plot(range(51), n_anly_plt[0],color = 'black', linewidth = 0.5, label = 'N = 1')

plt.plot(range(51), n_anly_plth[0], marker = '4',color = 'purple', linestyle = 'None', label = 'N = 3 Sim')
plt.plot(range(51), n_anly_plt[1],color = 'r', linewidth = 1.5, label = 'N = 3 Theo')

plt.plot(range(51), n_anly_plth[1], marker = '4',color = 'darkgreen', linestyle = 'None', label = 'N = 3 Sim')
plt.plot(range(51), n_anly_plt[2],color = 'g', linewidth = 1.5, label = 'N = 7 Theo')

plt.plot(range(51), n_anly_plth[2], marker = '4',color = 'cyan', linestyle = 'None', label = 'N = 3 Sim')
plt.plot(range(51), n_anly_plt[3],color = 'b', linewidth = 1.5, label = 'N = 11 Theo')

# Set the x and y pltis labels
plt.xlabel("Probabililty of codeword error")
plt.ylabel("Cross-over probability of 'p' of BSC")
plt.title('Performance of rate 1/N repetition codes over BSC(p) channels')

# Set the y-pltis limits to be between 0 and 0.5
plt.ylim([0, 0.5])

# Show the plot
plt.savefig('discrete_graph.png', dpi=1500)

plt.legend()
plt.show()

##################################################################

```
